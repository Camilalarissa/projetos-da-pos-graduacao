```python
# Caderno de Estudos Teóricos: Fundamentos de Análise de Dados

Este repositório armazena minhas anotações, formulações matemáticas e implementações práticas dos conceitos fundamentais de Análise e Ciência de Dados, servindo como base teórica para o desenvolvimento de soluções inteligentes.

##  Roteiro de Aprendizagem e Árvore Conceitual

### 1. Estatística Descritiva e Exploração de Dados (EDA)
- **Medidas de Tendência Central**: Média Aritmética, Média Ponderada, Mediana e Moda (identificação de assimetrias).
- **Medidas de Dispersão e Variabilidade**: Variância Amostral/Populacional, Desvio Padrão, Amplitude Interquartílica (IQR) e Coeficiente de Variação.
- **Análise de Distribuição**: Assimetria (Skewness), Curtose (Kurtosis) e a aplicação prática do Teorema do Limite Central.

### 2. Inferência Estatística e Testes de Hipóteses
- **Formulação de Hipóteses**: Definição clara de Hipótese Nula ($H_0$) e Hipótese Alternativa ($H_1$).
- **Erros de Decisão**: Entendimento visual do Erro Tipo I ($\alpha$) e Erro Tipo II ($\beta$).
- **Testes Paramétricos vs. Não-Paramétricos**: Aplicação de Teste t de Student, ANOVA, Teste qui-quadrado ($\chi^2$) e a interpretação matemática do p-valor.

### 3. Probabilidade e Modelagem Estocástica
- **Espaço Amostral e Eventos**: Teoremas fundamentais da probabilidade aditiva e multiplicativa.
- **Probabilidade Condicional**: Teorema de Bayes e suas aplicações na classificação de dados e inteligência artificial.
- **Distribuições Teóricas**: Modelagem de variáveis discretas e contínuas (Binomial, Poisson, Normal/Gaussiana).

### 4. Análise de Relacionamento entre Variáveis
- **Covariância e Correlação**: Entendimento geométrico do Coeficiente de Correlação de Pearson, Spearman e Kendall.
- **Introdução à Modelagem Preditiva**: Fundamentos matemáticos da Regressão Linear Simples, interpretação do $R^2$ (Coeficiente de Determinação) e análise de resíduos.
```
